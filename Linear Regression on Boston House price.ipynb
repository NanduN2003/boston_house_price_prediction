```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
### SPlitting
```

```python
##TrainTest Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_train.shape
X_train
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
X_test.shape
```

```python
#MOdel Training
from sklearn.linear_model import LinearRegression
rgr=LinearRegression()
```

```python
rgr.fit(X_train,y_train)
```

```python
##Print the coeficent and the intercwpt
```

```python
print(rgr.coef_)
```

```python
print(rgr.intercept_)
```

```python
##Prediction with test data
rgr_prediction=rgr.predict(X_test)
```

```python
rgr_prediction
```

```python
##plot scatter plot for prediction
plt.scatter(y_test,rgr_prediction)
```

```python
residuals=y_test-rgr_prediction
```

```python
residuals
```

```python
#Performance Metrics
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print(mean_absolute_error(y_test,rgr_prediction))
print(mean_squared_error(y_test,rgr_prediction))
```

```python
##R square and adjusted R square
from sklearn.metrics import r2_score
score=r2_score(y_test,rgr_prediction)
```
